**Load the Data:**

```python
import json
import requests

data = json.loads(requests.get('http://s3-us-west-2.amazonaws.com/boundaries.latimes.com/archive/1.0/boundary-set/la-county-neighborhoods-current.geojson').text)

from pandas import json_normalize
df = json_normalize(data["features"])

coords = 'geometry.coordinates'

fullDf = (df[coords].apply(lambda r: [(i[0],i[1]) for i in r[0]])
           .apply(pd.Series).stack()
           .reset_index(level=1).rename(columns={0:coords,"level_1":"point"})
           .join(df.drop(coords,1), how='left')).reset_index(level=0)

fullDf[['lat','long']] = fullDf[coords].apply(pd.Series)

fullDf
```

```python
def getRecommendedVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only needed information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    recommended_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    recommended_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(recommended_venues)
```

```python
LIMIT = 100 # limit of number of venues returned by Foursquare API

radius = 500 # define radius
```

```python
#Lunch time trending venues, function executed at ~1 pm LA time

rec_venues = getRecommendedVenues(names=centralDf['Neighborhood'],
                                   latitudes=centralDf['Latitude'],
                                   longitudes=centralDf['Longitude']
                                  )

print(rec_venues.shape)
rec_venues.head()
```

**Analyze each neighborhood to find out the 5 most common types of venues in each**

```python
# one hot encoding
rec_onehot = pd.get_dummies(rec_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column to dataframe (again)
rec_onehot['Neighborhood'] = rec_venues['Neighborhood'] 

# move neighborhood column to the first col
fixed_columns = [rec_onehot.columns[133]] + list(rec_onehot.columns[:])
rec_onehot = rec_onehot[fixed_columns]
rec_onehot = rec_onehot.loc[:,~rec_onehot.columns.duplicated()]

rec_onehot.head()
```

```python
rec_onehot.shape
```

```python
rec_grouped = rec_onehot.groupby('Neighborhood').mean().reset_index()
rec_grouped
```

**Now, here is each neighborhood along with its 5 most common recommended venue types**

```python
num_top_venues = 5

for nhood in rec_grouped['Neighborhood']:
    print("----"+nhood+"----")
    temp = rec_grouped[rec_grouped['Neighborhood'] == nhood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

**At a glance, we can see that Arlington Heights, Downtown, Hollywood, Harvard Heights, Silver Lake, and Larchmont all have restaurants highlighting their recommended venues with somewhat high frequencies. So, these 6 neighborhoods may be potential spots to open a new pizza place since it seems that resturants in these neighborhoods get good foot traffic. To get more info, we will do some clustering**

```python
def get_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

**Get 5 most common venues for each neighborhood**

```python
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = rec_grouped['Neighborhood']

for ind in np.arange(rec_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = get_most_common_venues(rec_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

**Setup for k means clustering**

```python
# set number of clusters
kclusters = 5

rec_grouped_clustering = rec_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(rec_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
rec_merged = centralDf
rec_merged = rec_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

rec_merged.head()
```
